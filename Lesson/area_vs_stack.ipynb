***single stackplot***

```python
import matplotlib.pyplot as plt 

x = [10,20,30,40,50,60,70]
area= [2,4,6,5,5,4,5]

plt.stackplot(x,area)

plt.legend()
plt.show()
```

***multiple stackplot***

```python
import matplotlib.pyplot as plt 

x = [10,20,30,40,50,60,70]
area= [2,4,6,5,5,4,5]
area2= [3,5,6,7,8,5,6]
area3= [2,6,7,8,9,9,8]
lbl = ["area","area2","area3"]

plt.stackplot(x,area,area2,area3,labels=lbl,colors= ["r","b","g"],baseline="sym")

plt.legend(loc=2)
plt.show()
```
